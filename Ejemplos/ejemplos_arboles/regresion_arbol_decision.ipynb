# Regresión de Arbol de Decisión

## Importar las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar el conjunto de datos

```python
datos = pd.read_csv('Position_Salaries.csv')
datos.head()
```

```python
X = datos.iloc[:, 1:-1].values
y = datos.iloc[:, -1].values
```

## Entrenar el modelo de Regresión de Arbol de Decisión con el conjunto de datos completo

**Nota:** Para ambos modelos de árboles de decisión y de Random Forest, no es necesario normalizar datos ya que no hay una fórmula a utilizar.  Simplemente se hacen divisiones (agrupaciones) sucesivas pero no ya operaciones entre variables.

```python
from sklearn.tree import DecisionTreeRegressor
regresor = DecisionTreeRegressor(random_state = 0)
regresor.fit(X, y)
```

## Predecir un nuevo resultado

```python
regresor.predict([[6.5]])
```

## Visualizar los resultados de la Regresión de Arbol de Decisión (mejor resolución)

```python
X_malla = np.arange(min(X), max(X), 0.01)
X_malla = X_malla.reshape((len(X_malla), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_malla, regresor.predict(X_malla), color = 'blue')
plt.title('Verdad o Mentira (Regresión de Arbol de Decisión)')
plt.xlabel('Nivel de Puesto')
plt.ylabel('Salario')
plt.show()
```

En este caso la predicción no es muy buena.  El algoritmo simplemente optó por dar el valor real (salario) de cada punto para todo el rango donde se encuentra el punto.  La razón es porque el modelo de árbol de decisión no es muy bueno cuando hay muy pocas variables independientes (features).  Este modelo "brilla" cuando hay muchas variables, ie. es multidimensional.
