# Regresión Random Forest

## Importar las librerías

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importar el conjunto de datos

```python
datos = pd.read_csv('Position_Salaries.csv')
datos.head(n=10)
```

```python
X = datos.iloc[:, 1:-1].values
y = datos.iloc[:, -1].values
```

## Entrenar el modelo de Regresión Random Forest con el conjunto de datos completo

```python
from sklearn.ensemble import RandomForestRegressor
regresor = RandomForestRegressor(n_estimators = 500, random_state = 0)
regresor.fit(X, y)
```

n_estimators = cantidad de árboles

## Predicción de un nuevo resultado

```python
regresor.predict([[6.5]])
```

## Visualizar los resultados de la Regresión de Random Forest (resolución mayor)

```python
X_malla = np.arange(min(X), max(X), 0.01)
X_malla = X_malla.reshape((len(X_malla), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_malla, regresor.predict(X_malla), color = 'blue')
plt.title('Verdad o Mentira (Regresión Random Forest)')
```

Vemos que hay una mejor resolución que con el modelo de árbol de decisión simple, esto es porque hay más árboles haciendo el cálculo.

Es importante ver que agregar más árboles no necesariamente va a generar más gradas.  Esto  se puede ver haciendo el mismo ejercicio pero con 100, o 500 árboles.  Lo que se ve es que no hay mucho cambio en el número de gradas pero se están moviendo para ajustarse a una mejor resolución.
